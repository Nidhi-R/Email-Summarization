```python
import nltk
import numpy as np
from nltk.tokenize import sent_tokenize, word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models import Word2Vec
from transformers import pipeline
```

```python
nltk.download('punkt')

# Function to preprocess text
def preprocess_text(text):
    sentences = sent_tokenize(text)
    tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in sentences]
    return sentences, tokenized_sentences

def extractive_summarize_text(sentences, tokenized_sentences, num_sentences=2):
    # Load and train Word2Vec model
    model = Word2Vec(sentences=tokenized_sentences, vector_size=300, window=5, min_count=1, workers=4)

    # Generate sentence vectors
    sentence_vectors = []
    for sentence in tokenized_sentences:
        vectors = [model.wv[word] for word in sentence if word in model.wv]
        if vectors:
            sentence_vector = np.mean(vectors, axis=0)
            sentence_vectors.append(sentence_vector)

    # Calculate similarity matrix
    similarity_matrix = cosine_similarity(sentence_vectors)

    # Rank sentences based on similarity to identify top sentences for summary
    sentence_scores = np.sum(similarity_matrix, axis=1)
    ranked_sentences = sorted(range(len(sentence_scores)), key=lambda i: sentence_scores[i], reverse=True)[:num_sentences]

    # Generate the extractive summary
    extractive_summary = ' '.join([sentences[i] for i in ranked_sentences])
    return extractive_summary
# Naive summarization without using word embeddings
def naive_summarize_text(sentences, num_sentences=1):
    # Simply select the first num_sentences sentences
    naive_summary = ' '.join(sentences[:num_sentences])
    return naive_summary

# Example text
input_text = """Most modern deep learning models are based on multi-layered artificial neural networks such as convolutional neural networks and transformers, although they can also include propositional formulas or latent variables organized layer-wise in deep generative models such as the nodes in deep belief networks and deep Boltzmann machines.In deep learning, each level learns to transform its input data into a slightly more abstract and composite representation. In an image recognition application, the raw input may be a matrix of pixels; the first representational layer may abstract the pixels and encode edges; the second layer may compose and encode arrangements of edges; the third layer may encode a nose and eyes; and the fourth layer may recognize that the image contains a face.
Importantly, a deep learning process can learn which features to optimally place in which level on its own. This does not eliminate the need for hand-tuning; for example, varying numbers of layers and layer sizes can provide different degrees of abstraction.Deep learning algorithms can be applied to unsupervised learning tasks. This is an important benefit because unlabeled data are more abundant than the labeled data. Examples of deep structures that can be trained in an unsupervised manner are deep belief networks.[12][17]
Machine learning models are now adept at identifying complex patterns in financial market data. Due to the benefits of artificial intelligence, investors are increasingly utilizing deep learning techniques to forecast and analyze trends in stock and foreign exchange markets.
"""
# Preprocess the text
sentences, tokenized_sentences = preprocess_text(input_text)
```

```python
# Extractive summarization using word embeddings
extractive_summary = extractive_summarize_text(sentences, tokenized_sentences)
print("\nExtractive Summary:")
print(extractive_summary)

# Naive summarization without using word embeddings
naive_summary = naive_summarize_text(sentences)
print("\nNaive Summary:")
print(naive_summary)
```

```python
from transformers import BertTokenizer, BertModel
import torch

# Function to preprocess text using BERT tokenizer
def preprocess_text_bert(text):
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    sentences = sent_tokenize(text)
    tokenized_sentences = [tokenizer.tokenize(sentence.lower()) for sentence in sentences]
    return sentences, tokenized_sentences, tokenizer

# Function to extract BERT embeddings for sentences
def get_bert_embeddings(tokenized_sentences, tokenizer):
    model = BertModel.from_pretrained('bert-base-uncased')
    model.eval()

    max_length = max(len(sentence) for sentence in tokenized_sentences)

    # Flatten and pad the tokenized sentences
    flat_tokenized_sentences = [tokenizer.convert_tokens_to_ids(sentence) + [tokenizer.pad_token_id] * (max_length - len(sentence)) for sentence in tokenized_sentences]

    # Create attention masks
    attention_masks = [[1] * len(sentence) + [0] * (max_length - len(sentence)) for sentence in tokenized_sentences]

    embeddings = []
    with torch.no_grad():
        for sentence, attention_mask in zip(flat_tokenized_sentences, attention_masks):
            inputs = {"input_ids": torch.tensor([sentence]), "attention_mask": torch.tensor([attention_mask])}
            outputs = model(**inputs)
            embeddings.append(outputs.last_hidden_state.mean(dim=1).squeeze().numpy())

    return embeddings

# Function to perform extractive summarization using BERT embeddings
def bert_extractive_summarize_text(sentences, tokenized_sentences, tokenizer, num_sentences=2):
    embeddings = get_bert_embeddings(tokenized_sentences, tokenizer)

    # Calculate similarity matrix
    similarity_matrix = cosine_similarity(embeddings)

    # Rank sentences based on similarity to identify top sentences for summary
    sentence_scores = np.sum(similarity_matrix, axis=1)
    ranked_sentences = sorted(range(len(sentence_scores)), key=lambda i: sentence_scores[i], reverse=True)[:num_sentences]

    # Generate the extractive summary
    extractive_summary = ' '.join([sentences[i] for i in ranked_sentences])
    return extractive_summary

# Preprocess the text using BERT tokenizer
sentences_bert, tokenized_sentences_bert, tokenizer = preprocess_text_bert(input_text)

# Extractive summarization using BERT embeddings
bert_extractive_summary = bert_extractive_summarize_text(sentences_bert, tokenized_sentences_bert, tokenizer)
print("\nBERT Extractive Summary:")
print(bert_extractive_summary)
```

```python
import spacy
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import sent_tokenize, word_tokenize

# Download the spaCy model with GloVe embeddings
nlp = spacy.load("en_core_web_md")

def preprocess_text_glove(text):
    sentences = sent_tokenize(text)
    tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in sentences]
    return sentences, tokenized_sentences

def get_glove_embeddings(tokenized_sentences):
    embeddings = []
    for sentence in tokenized_sentences:
        # Calculate the sentence vector by averaging word vectors
        vectors = [nlp(word).vector for word in sentence]
        if vectors:
            sentence_vector = np.mean(vectors, axis=0)
            embeddings.append(sentence_vector)
    return embeddings

def glove_extractive_summarize_text(sentences, tokenized_sentences, num_sentences=2):
    embeddings = get_glove_embeddings(tokenized_sentences)

    # Calculate similarity matrix
    similarity_matrix = cosine_similarity(embeddings)

    # Rank sentences based on similarity to identify top sentences for summary
    sentence_scores = np.sum(similarity_matrix, axis=1)
    ranked_sentences = sorted(range(len(sentence_scores)), key=lambda i: sentence_scores[i], reverse=True)[:num_sentences]

    # Generate the extractive summary
    extractive_summary = ' '.join([sentences[i] for i in ranked_sentences])
    return extractive_summary

# Preprocess the text using GloVe vectors
sentences_glove, tokenized_sentences_glove = preprocess_text_glove(input_text)

# Extractive summarization using GloVe embeddings
glove_extractive_summary = glove_extractive_summarize_text(sentences_glove, tokenized_sentences_glove)
print("\nGloVe Extractive Summary:")
print(glove_extractive_summary)


```

```python
from rouge import Rouge

def calculate_rouge(reference, candidate):
    rouge = Rouge()
    scores = rouge.get_scores(candidate, reference)
    return scores[0]

# Original reference summary
reference_summary = """Deep learning models, based on multi-layered artificial neural networks, transform input data into abstract representations. They can learn which features to place in each level, but still require hand-tuning. Deep learning can be applied to unsupervised learning tasks, as unlabeled data is more abundant. Examples of deep structures that can be trained in an unsupervised manner include deep belief networks. Machine learning models are increasingly used by investors to forecast and analyze trends in stock and foreign exchange markets due to the benefits of artificial intelligence.
"""
# BERT Extractive Summary
bert_rouge_scores = calculate_rouge(reference_summary, bert_extractive_summary)
print("BERT Rouge Scores:", bert_rouge_scores)

# GloVe Extractive Summary
glove_rouge_scores = calculate_rouge(reference_summary, glove_extractive_summary)
print("\nGloVe Rouge Scores:", glove_rouge_scores)

# Word2Vec Extractive Summary (Assuming you have the `extractive_summary` function for Word2Vec)
word2vec_extractive_summary = extractive_summarize_text(sentences, tokenized_sentences)
word2vec_rouge_scores = calculate_rouge(reference_summary, word2vec_extractive_summary)
print("\nWord2Vec Rouge Scores:", word2vec_rouge_scores)

# Naive Summary
naive_rouge_scores = calculate_rouge(reference_summary, naive_summary)
print("\nNaive Rouge Scores:", naive_rouge_scores)
```
